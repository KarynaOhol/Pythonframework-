```python

from math import sqrt 

def is_not_prime(n):
    ans = False
    for i in range(2, int(sqrt(n)) + 1):
        if n % i == 0:
            ans = True
    return ans
 
def sortedArray(arr, n) : 
    v = [] 
    for i in range(n) : 
        if is_not_prime(arr[i]) : 
            v.append(arr[i]) 
  
     
    v.sort(reverse = True) 
    print(v)
  
  
# Driver code 
if __name__ == "__main__" : 
  
  try:  
    arr = []
    n = int(input("Enter number of elements : ")) 
      
    for i in range(0, n): 
      num = int(input()) 
      arr.append(num)

    sortedArray(arr, n) 
  except ValueError as ve:
        print("Input wrong value: must be positive int numbers ")
```
